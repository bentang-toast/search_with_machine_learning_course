```python
df_merged = df.merge(df_count, how='left', on='category').merge(parents_df, how='left', on='category')
df_merged
```

```python
df_merged['category'] = np.where(df_merged['count'] < min_queries, df_merged['parent'], df_merged['category'])
df_merged
```

```python
df_merged['category'] = np.where(df_merged['count'] < 1000, df_merged['parent'], df_merged['category'])
df_merged
```

```python
df = df_merged
df
```

```python
# Create labels in fastText format.
df['label'] = '__label__' + df['category']
df
```
